```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

sinasc = pd.read_csv('Data/SINASC_RO_2019.csv')
```

#### **2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Lista de variáveis categóricas para plotar
variaveis = ['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE']


# Configurar o tamanho da figura
plt.figure(figsize=(10, 20))

# Criar os gráficos de barras
for i, var in enumerate(variaveis, 1):
    plt.subplot(len(variaveis), 1, i)
    sns.countplot(data=sinasc, x=var, palette='viridis')
    plt.title(f'Distribuição de {var}', fontsize=14)
    plt.xlabel('')
    plt.ylabel('')
    plt.box(False)  # Remove as bordas

# Ajustar layout
plt.tight_layout()
plt.show()
```

#### **3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS**

```python
# Agrupar os dados por 'CONSULTAS' e calcular a média do 'APGAR5'
apgar5_por_consultas = sinasc.groupby('CONSULTAS')['APGAR5'].mean().reset_index()

# Criar o gráfico de barras
plt.figure(figsize=(10, 6))
sns.barplot(x='CONSULTAS', y='APGAR5', data=apgar5_por_consultas, palette='viridis')
plt.title('APGAR5 Médio por Número de Consultas Pré-Natais', fontsize=16)
plt.xlabel('Número de Consultas Pré-Natais', fontsize=12)
plt.ylabel('APGAR5 Médio', fontsize=12)
plt.xticks(rotation=45, ha='right')  # Rotacionar os rótulos do eixo x para melhor legibilidade
plt.show()
```

#### **4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS**

```python
# Lista de variáveis a serem plotadas
variaveis = ['SEXO', 'GESTACAO', 'ESTCIVMAE', 'CONSULTAS']

# Configurar o tamanho da figura
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

# Criar os gráficos de barras
for i, var in enumerate(variaveis):
    sns.countplot(data=sinasc, x=var, palette='viridis', ax=axes[i])
    axes[i].set_title(f'Frequência de {var}', fontsize=14)
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Frequência')

    # Rotacionar rótulos do eixo X se necessário
    axes[i].tick_params(axis='x', rotation=45)

# Ajustar layout
plt.tight_layout()
plt.show()
```
